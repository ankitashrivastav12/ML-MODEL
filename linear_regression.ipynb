```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python

plt.xlabel("area(sq ft)")
plt.ylabel("price(US $)")
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
ref=linear_model.LinearRegression()
ref.fit(df[["area"]],df.price)
```

```python
ref.intercept_
```

```python
ref.predict([[5000]])
```

```python
ref.coef_
```

```python
df1=pd.read_csv("areas.csv")
df1
```

```python
p=ref.predict(df1)
df1["PRICE"]=p
df1
```

```python
df1.to_csv("prediction.csv")
```

```python
plt.xlabel("area(sq ft)")
plt.ylabel("price(US $)")
plt.scatter(df.area,df.price,color="red",marker="+")
plt.plot(df.area,ref.predict(df[["area"]]),color="green")
```
