```python
from sklearn.model_selection import cross_val_score
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df["target"]=iris.target
df["target"]=df.target.apply(lambda x:iris.target_names[x])
df
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(),iris.data,df.target,cv=5)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(SVC(),iris.data,df.target)
```

```python
from sklearn.model_selection import cross_val_score
s1=cross_val_score(RandomForestClassifier(),iris.data,df.target,cv=5)
np.average(s1)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(RandomForestClassifier(),iris.data,df.target,cv=5)
```
