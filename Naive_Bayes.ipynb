```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df
```

```python

inputs=df.drop(["PassengerId","Survived","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns")
target=df.Survived
```

```python
inputs
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies
```

```python
final=pd.concat([inputs,dummies],axis="columns")
final=final.drop(["Sex","male"],axis="columns")
final
```

```python
final.isna().sum()
```

```python
final.Age=final.Age.fillna(final.Age.mean())
final
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(final,target,test_size=0.2,random_state=1,stratify=target)
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
```

```python
gb.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),x_train,y_train)
```

```python
gb.score(x_test,y_test)
```

```python
gb.predict_proba(x_test[:10])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_train_count=cv.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
emails=["Had your mobile 11 months or more? U R entitled to Update to the latest colour mobiles with camera for Free! Call The Mobile Update Co FREE on 08002986030",
        "Fine if thats the way u feel. Thats the way its gota b"]
emails_count=cv.transform(emails)
model.predict(emails_count)
x_test_count=cv.transform(x_test)
model.score(x_test_count,y_test)
```
