```python
import pandas as pd
df=pd.read_csv("heart.csv")
df
```

```python
x=df.drop(["HeartDisease"],axis="columns")
y=df.HeartDisease
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le_cpt=LabelEncoder()
le_sex=LabelEncoder()
le_restecg=LabelEncoder()
le_exagin=LabelEncoder()
le_slope=LabelEncoder()
x["chest_pain_type"]=le_cpt.fit_transform(x[["ChestPainType"]])
x["gender"]=le_sex.fit_transform(x[["Sex"]])
x["Resting_ECG"]=le_sex.fit_transform(x[["RestingECG"]])
x["Exercise_Angina"]=le_sex.fit_transform(x[["ExerciseAngina"]])
x["st_Slope"]=le_sex.fit_transform(x[["ST_Slope"]])
inputs=x.drop(["ChestPainType","Sex","ExerciseAngina","RestingECG","ST_Slope"],axis="columns")
inputs
```

```python
inputs.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(inputs,y)
model.predict([[40,140,289,0,172,0.0,1,1,1,0,2]])
model.score(inputs,y)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(inputs,y)
model.predict([[40,140,289,0,172,0.0,1,1,1,0,2]])
model.score(inputs,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(inputs,y)
model.predict([[40,140,289,0,172,0.0,1,1,1,0,2]])
model.score(inputs,y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaler=scaler.fit_transform(inputs)
x_scaler
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.97)
x_pca=pca.fit_transform(inputs,y)
x_pca
```
