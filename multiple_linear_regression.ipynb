```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import math
from word2number import w2n
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
med_bed=math.floor(df.bedrooms.median())
med_bed
```

```python
df.bedrooms=df.bedrooms.fillna(med_bed)
df
```

```python
ref=linear_model.LinearRegression()
ref.fit(df[["area","bedrooms","age"]],df.price)
```

```python
ref.predict([[3800,2,29]])
```

```python
df1=pd.read_csv("hiring.csv")
df1
```

```python
df1.experience=df1.experience.fillna("zero")
df1
```

```python
med_score=math.floor(df1.test_score.mean())
df1.test_score=df1.test_score.fillna(med_score)
df1
```

```python
df1.experience = df1.experience.astype(str)
df1.experience = df1.experience.apply(w2n.word_to_num)
df1
```

```python
ref=linear_model.LinearRegression()
ref.fit(df1[["experience","test_score","interview_score"]],df1.salary)
```

```python
ref.coef_
```

```python
ref.intercept_
```

```python
ref.predict([[2,9,6]])
```
